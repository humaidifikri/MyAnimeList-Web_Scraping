```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://myanimelist.net/anime/genre/9/Ecchi'
page = requests.get(url).text
html_soup = BeautifulSoup(page,'html.parser')
list_anime = html_soup.find('div',{'class':'seasonal-anime-list'})
list_anime
```

```python
anime_names = list_anime.findAll('a',{'class':'link-title'})
anime_names
```

```python
anime_names_in_list = [name.get_text() for name in anime_names]
print(anime_names_in_list)
```

```python
anime_ratings = list_anime.findAll('div',{'class':'scormem-container'})
anime_ratings
```

```python
anime_rating_list = [rating.get_text().split('\n')[2] for rating in anime_ratings]
print(anime_rating_list)
```

```python
anime_infos = list_anime.findAll('div',{'class':'info'})
anime_infos
```

```python
anime_info_list = [anime_info.get_text()[:8] for anime_info in anime_infos]
print(anime_info_list)
```

```python
categories=[]
years=[]
for item_list in anime_info_list:
    category, year = item_list.split(', ')
    categories.append(category)
    years.append(year)
    
print(categories,years)
```

```python
anime_status = list_anime.find_all('span',{'class':'item finished'})
status = [i.get_text() for i in anime_status]
print(status)
```

```python
anime_eps = [eps.get_text()[17:23] for eps in anime_infos]
print(anime_eps)
```

```python
df = pd.DataFrame({'anime_titles':anime_names_in_list,
                  'categories':categories,
                  'years':years,
                  'ratings':anime_rating_list,
                  'status':anime_status,
                  'total_episode':anime_eps})
df
```
